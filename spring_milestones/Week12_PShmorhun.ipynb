# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

### Graph 1: Side-By-Side Plots
**Advantages:**
- Clear separation of each week's data; its easy to see the individual trends for both weeks between `time_series_0` and `time_series_1`.
- Helps avoid confusion if the data for each week is noisy or highly variable.
**Disadvantages:**
- Harder to compare values between weeks. The viewer must visually match points across two seprate plots.
- Takes up more visual space in a presentation or paper.

### Graph 2: Overlayed Line Plot
**Advantages:** 
- Its easy to see how the data from `time_series_0` and `time_series_1` compare to one another on each day of the week.
- More compact, taking up less space in a presentation or paper.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data
data = {
    'Product A': [400, 420, 430, 400, 300, 270, 250],
    'Product B': [370, 400, 410, 380, 290, 260, 240],
    'Product C': [0,   0,  100, 180, 200, 240, 180],
    'Product D': [0,   0,    0, 160, 270, 220, 210],
    'Product E': [0,   0,    0,   0,   0, 100, 210],
}
years = ['2008', '2009', '2010', '2011', '2012', '2013', '2014']
df = pd.DataFrame(data, index=years)

# Plot setup
bar_width = 0.1
x = np.arange(len(df.columns))  
colors = ['royalblue', 'brown', 'olive', 'purple', 'skyblue', 'darkorange', 'steelblue']

plt.figure(figsize=(10, 5))

# Plot each year's data as a set of bars
for i, year in enumerate(df.index):
    plt.bar(x + i * bar_width, df.loc[year], width=bar_width, label=year, color=colors[i])

# Formatting
plt.xticks(x + bar_width * 3, df.columns)
plt.yticks(np.arange(0, 501, 100), [f"${v}" for v in range(0, 501, 100)])
plt.ylim(0, 500)
plt.title("Average Retail Product Price Per Year", fontsize=14, weight='bold')
plt.legend(title="Year", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
